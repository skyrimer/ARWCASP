```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
street = gpd.read_parquet("processed_data/street.parquet")
search = gpd.read_parquet("processed_data/search.parquet")
outcomes = gpd.read_parquet("processed_data/outcomes.parquet")
```

## Street

```python
# Group by LSOA code and period, then count the number of crimes
crime_counts_street_by_LSOA = street.groupby(["LSOA code", "period"])["Crime ID"].count().reset_index()

# Rename for clarity
crime_counts_street_by_LSOA.rename(columns={"Crime ID": "Crime Count"}, inplace=True)

# Optional: sort for easier reading
crime_counts_street_by_LSOA.sort_values(by=["LSOA code", "period"], inplace=True)

# Final result: crimes per LSOA per year-month
print(crime_counts_street_by_LSOA)
```

#### Example, to see if it works (which it does :D)

```python

# Filter for specific LSOA code
lsoa_data = crime_counts_street_by_LSOA[crime_counts_street_by_LSOA['LSOA code'] == 'E01000001'].copy()

# Convert Period to Timestamp for plotting
lsoa_data['period'] = lsoa_data['period'].dt.to_timestamp()

# Plot
plt.figure(figsize=(10, 6))
plt.plot(lsoa_data['period'], lsoa_data['Crime Count'], marker='o')

# Styling
plt.title('Crime Trend for LSOA E01000001')
plt.xlabel('Period')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

plt.show()
```
